```python
import csv
import random
from faker import Faker

# Initialize Faker
fake = Faker()

# Define record counts
NUM_CUSTOMERS = 3400
NUM_ACCOUNTS = 5000
NUM_TRANSACTIONS = 10000
NUM_BRANCHES = 100
NUM_LOANS = 1600

# Generate Customers
with open('customers.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['customer_id', 'first_name', 'last_name', 'email', 'phone', 'address', 'date_of_birth'])
    for customer_id in range(1, NUM_CUSTOMERS + 1):
        writer.writerow([
            customer_id,
            fake.first_name(),
            fake.last_name(),
            fake.email(),
            fake.phone_number(),
            fake.address().replace('\n', ', '),
            fake.date_of_birth(minimum_age=18, maximum_age=75)
        ])

# Generate Branches
with open('branches.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['branch_id', 'branch_name', 'branch_address', 'branch_phone'])
    for branch_id in range(1, NUM_BRANCHES + 1):
        writer.writerow([
            branch_id,
            f"{fake.city()} Branch",
            fake.address().replace('\n', ', '),
            fake.phone_number()
        ])

# Generate Accounts
with open('accounts.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['account_id', 'customer_id', 'account_type', 'balance', 'opened_date'])
    account_types = ['Savings', 'Checking']
    for account_id in range(1, NUM_ACCOUNTS + 1):
        writer.writerow([
            account_id,
            random.randint(1, NUM_CUSTOMERS),
            random.choice(account_types),
            round(random.uniform(1000, 50000), 2),
            fake.date_between(start_date='-5y', end_date='today')
        ])

# Generate Transactions
with open('transactions.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['transaction_id', 'account_id', 'transaction_type', 'amount', 'transaction_date', 'description'])
    transaction_types = ['Deposit', 'Withdrawal', 'Transfer']
    for transaction_id in range(1, NUM_TRANSACTIONS + 1):
        writer.writerow([
            transaction_id,
            random.randint(1, NUM_ACCOUNTS),
            random.choice(transaction_types),
            round(random.uniform(10, 5000), 2),
            fake.date_between(start_date='-2y', end_date='today'),
            fake.sentence(nb_words=5)
        ])

# Generate Loans
with open('loans.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['loan_id', 'customer_id', 'loan_amount', 'interest_rate', 'loan_start_date', 'loan_end_date'])
    for loan_id in range(1, NUM_LOANS + 1):
        start_date = fake.date_between(start_date='-5y', end_date='-1y')
        end_date = fake.date_between(start_date=start_date, end_date='+5y')
        writer.writerow([
            loan_id,
            random.randint(1, NUM_CUSTOMERS),
            round(random.uniform(5000, 100000), 2),
            round(random.uniform(1.5, 5.0), 2),
            start_date,
            end_date
        ])
```
